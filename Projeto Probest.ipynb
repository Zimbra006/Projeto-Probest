```python
import numpy as np
import pandas as pd
import math
```

```python
df_chrome = pd.read_csv(r'C:\UFRJ\Materias\Probest\dataset_chromecast.csv')
df_chrome.head()
```

```python
df_tv = pd.read_csv(r'C:\UFRJ\Materias\Probest\dataset_smart-tv.csv')
df_tv.head()
```

```python
# Aplicação do log como pedido pela professora
def log10(num):
    return math.log(num + 1, 10)
```

```python
df_chrome['bytes_up'] = df_chrome['bytes_up'].map(lambda value: log10(value))
df_chrome['bytes_down'] = df_chrome['bytes_down'].map(lambda value: log10(value))

df_chrome.head()
```

```python
df_tv['bytes_up'] = df_tv['bytes_up'].map(lambda value: log10(value))
df_tv['bytes_down'] = df_tv['bytes_down'].map(lambda value: log10(value))

df_tv.head()
```

HISTOGRAMAS SEM HORÁRIO

```python
numBinsChrome = round(1 + 3.322 * math.log(df_chrome['bytes_up'].size, 10))
print(numBinsChrome)
df_chrome.hist(column = 'bytes_up', bins = numBinsChrome)
df_chrome.hist(column = 'bytes_down', bins = numBinsChrome)
```

```python
numBinsTv = round(1 + 3.322 * math.log(df_tv['bytes_up'].size, 10))
print(numBinsTv)
df_tv.hist(column = 'bytes_up', bins = numBinsTv)
df_tv.hist(column = 'bytes_down', bins = numBinsTv)
```
