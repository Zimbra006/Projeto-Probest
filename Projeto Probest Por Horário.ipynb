```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
from scipy.stats import norm, probplot
import statistics
```

```python
# Datasets
df_chrome = pd.read_csv(r'C:\UFRJ\Materias\Probest\dataset_chromecast.csv')
df_tv = pd.read_csv(r'C:\UFRJ\Materias\Probest\dataset_smart-tv.csv')

# Definição da função log pedida
def log10(num):
    return math.log(num + 1, 10)

# Changing the columns to the same scale
df_chrome['bytes_up'] = df_chrome['bytes_up'].map(lambda value: log10(value))
df_chrome['bytes_down'] = df_chrome['bytes_down'].map(lambda value: log10(value))

df_tv['bytes_up'] = df_tv['bytes_up'].map(lambda value: log10(value))
df_tv['bytes_down'] = df_tv['bytes_down'].map(lambda value: log10(value))
```

```python
def encontrarMaiorMediaEMediana(dataframe):
    maior_media, i_media = 0.0, 0
    maior_mediana, i_mediana = 0.0, 0
    
    for i in range(23):
        if dataframe['mean'][i] > maior_media:
            maior_media = dataframe['mean'][i]
            i_media = i
        if dataframe['median'][i] > maior_mediana:
            maior_mediana = dataframe['median'][i]
            i_mediana = i
    
    print('Maior média ' + str(maior_media) + ' encontrada na hora ' + str(i_media))
    print('Maior mediana ' + str(maior_mediana) + ' encontrada na hora ' + str(i_mediana))
        
```

```python
def MLEGaussian(dataset):
    size = dataset.size
    
    mean = dataset.sum() / size
    
    sigma_sum = 0
    for i in range(size):
        sigma_sum += (dataset[i] - mean)**2
        
    sigma = (sigma_sum / size)**(1/2)
    
    return (mean, sigma)
```
